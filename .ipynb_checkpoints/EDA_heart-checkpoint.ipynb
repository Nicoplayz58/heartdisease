##  Introducci贸n

La enfermedad card铆aca es una de las principales causas de muerte a nivel mundial. Su diagn贸stico temprano y preciso es fundamental para reducir la mortalidad y mejorar la calidad de vida de los pacientes. En este proyecto, realizamos un an谩lisis exploratorio detallado del dataset de enfermedades card铆acas proveniente de UCI (University of California, Irvine), con el objetivo de identificar las variables m谩s relevantes asociadas al diagn贸stico de la condici贸n.

Este an谩lisis incluye:

- Exploraci贸n univariada y bivariada de variables num茅ricas y categ贸ricas.
- Evaluaci贸n de distribuci贸n y normalidad de variables continuas.
- An谩lisis de correlaci贸n con la variable objetivo (`target`), que indica presencia (`1`) o ausencia (`0`) de enfermedad.
- Estudio de multicolinealidad entre variables predictoras.
- Evaluaci贸n estad铆stica de asociaci贸n entre variables categ贸ricas.
- Estimaci贸n de la importancia de caracter铆sticas mediante un modelo de Random Forest.
- Aplicaci贸n del modelo Random Forest para realizar predicciones sobre la presencia de enfermedad.

El prop贸sito es no solo identificar los factores que m谩s influyen en el diagn贸stico, sino tambi茅n sentar las bases para futuros modelos de clasificaci贸n que permitan predecir el riesgo card铆aco con mayor precisi贸n y respaldo estad铆stico.

## Ь Descripci贸n de Variables

El dataset contiene variables cl铆nicas, demogr谩ficas y de resultados de pruebas m茅dicas que pueden influir en el diagn贸stico de enfermedad card铆aca. A continuaci贸n, se describen las principales variables incluidas:

###  Variables Num茅ricas

- `age`: Edad del paciente (en a帽os).
- `trestbps`: Presi贸n arterial en reposo (en mm Hg).
- `chol`: Nivel de colesterol s茅rico en mg/dl.
- `thalach`: Frecuencia card铆aca m谩xima alcanzada.
- `oldpeak`: Depresi贸n del segmento ST inducida por el ejercicio (comparada con el reposo).

###  Variables Categ贸ricas

- `sex`: Sexo del paciente (0 = mujer, 1 = hombre).
- `cp`: Tipo de dolor en el pecho (0 = angina t铆pica, 1 = angina at铆pica, 2 = dolor no anginoso, 3 = asintom谩tico).
- `fbs`: Glucemia en ayunas > 120 mg/dl (1 = s铆, 0 = no).
- `restecg`: Resultados del electrocardiograma en reposo (0, 1, 2).
- `exang`: Angina inducida por ejercicio (1 = s铆, 0 = no).
- `slope`: Pendiente del segmento ST durante el esfuerzo (0 = descendente, 1 = plana, 2 = ascendente).
- `ca`: N煤mero de vasos principales coloreados por fluoroscop铆a (0 a 3).
- `thal`: Tipo de talasemia (1 = normal, 2 = defecto fijo, 3 = defecto reversible).

###  Variable Objetivo

- `target`: Diagn贸stico de enfermedad card铆aca (1 = presencia, 0 = ausencia).

```python
from scipy.stats import shapiro
import pandas as pd

# Variables num茅ricas del dataset
numerical_vars = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']

# Lista para guardar resultados
shapiro_results = []

# Aplicar Shapiro-Wilk a cada variable
for col in numerical_vars:
    stat, p = shapiro(df[col])
    shapiro_results.append({
        "Variable": col,
        "Estad铆stico Shapiro-Wilk": stat,
        "p-valor": p,
        "驴Distribuci贸n Normal?": "S铆" if p > 0.05 else "No"
    })

# Mostrar resultados como DataFrame
shapiro_df = pd.DataFrame(shapiro_results)
display(shapiro_df)
```

```python
# A帽adir la columna "target" al an谩lisis de correlaci贸n Spearman
vars_with_target = numerical_vars + ["target"]
spearman_corr_with_target = df[vars_with_target].corr(method='spearman').round(2)

# Mostrar la nueva matriz incluyendo el target
spearman_corr_with_target
```

```python
import plotly.express as px

fig_heatmap = px.imshow(spearman_corr_with_target,
                        text_auto=True,
                        color_continuous_scale="RdBu_r",  # Escala rojo-azul
                        zmin=-1, zmax=1,
                        title="Matriz de Correlaci贸n Spearman (incluyendo target)",
                        labels=dict(color="Correlaci贸n"))
fig_heatmap.update_layout(template="plotly_dark",
                          width=600, height=600)
fig_heatmap.show()
```

```python
import pandas as pd
from scipy.stats import chi2_contingency
import numpy as np

# Lista de variables categ贸ricas en el dataset
cat_vars = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal']

# Funci贸n para calcular Cram茅r's V
def cramers_v(confusion_matrix):
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    return np.sqrt(phi2 / min(k - 1, r - 1))

# Evaluar cada variable categ贸rica contra target
association_results = []
for col in cat_vars:
    confusion_matrix = pd.crosstab(df[col], df["target"])
    chi2, p, _, _ = chi2_contingency(confusion_matrix)
    cramers = cramers_v(confusion_matrix)
    association_results.append({
        "Variable": col,
        "Chi虏": round(chi2, 2),
        "p-valor": round(p, 5),
        "Cram茅r's V": round(cramers, 3),
        "驴Asociaci贸n significativa (p < 0.05)?": "S铆" if p < 0.05 else "No"
    })

association_df = pd.DataFrame(association_results)

print("\nAsociaci贸n entre variables categ贸ricas y `target` (Chi虏 + Cram茅r's V):\n")
print(association_df.to_string(index=False))
```

```python
# A帽adir target a la matriz de Cram茅r's V
cramers_matrix_with_target = cramers_matrix.copy()

# Calcular Cram茅r's V entre target y todas las categ贸ricas
target_cramers = []
for var in cat_vars:
    confusion_matrix = pd.crosstab(df[var], df["target"])
    cramers = cramers_v(confusion_matrix)
    target_cramers.append(round(cramers, 3))

# Agregar columna y fila de target
cramers_matrix_with_target["target"] = target_cramers
cramers_matrix_with_target.loc["target"] = target_cramers + [1.0]  # autocorrelaci贸n target-target

cramers_matrix_with_target
```

```python
import plotly.express as px

fig_cramers = px.imshow(cramers_matrix_with_target.astype(float),
                        color_continuous_scale="RdBu_r",
                        zmin=0, zmax=1,
                        title="Matriz de Asociaci贸n Categ贸rica (Cram茅r's V) incluyendo target",
                        labels=dict(color="Cram茅r's V"))

# Si text_auto no funciona, simplemente actualiza el layout
fig_cramers.update_traces(texttemplate="%{z:.2f}", textfont_size=12)

fig_cramers.update_layout(template="plotly_dark",
                          width=700, height=700)
fig_cramers.show()
```

##  Conclusiones del An谩lisis Descriptivo

Tras realizar un an谩lisis exploratorio y estad铆stico del dataset de enfermedades card铆acas, se obtuvieron los siguientes hallazgos relevantes:

### 1. Distribuci贸n de la muestra
- La clase objetivo (`target`) est谩 relativamente balanceada, lo que permite aplicar modelos de clasificaci贸n sin requerir t茅cnicas agresivas de balanceo.
- Se observ贸 una mayor prevalencia de la enfermedad en pacientes con caracter铆sticas cl铆nicas espec铆ficas como depresi贸n del ST elevada (`oldpeak`) y frecuencia card铆aca m谩xima baja (`thalach`).

### 2. Variables num茅ricas
- Ninguna de las variables num茅ricas sigue una distribuci贸n normal seg煤n la prueba de Shapiro-Wilk, por lo que se utilizaron correlaciones de Spearman.
- Las variables m谩s correlacionadas con la presencia de enfermedad fueron:
  - `thalach` ( = 0.43)
  - `oldpeak` ( = -0.42)
  - `age` ( = -0.24)

### 3. Variables categ贸ricas
- La prueba de Chi-cuadrado y el coeficiente de Cram茅rs V indicaron una fuerte asociaci贸n entre la enfermedad y las siguientes variables categ贸ricas:
  - `thal` (V = 0.531)
  - `cp` (V = 0.519)
  - `ca` (V = 0.495)
  - `exang` (V = 0.430)
- Otras variables como `fbs` mostraron una relaci贸n estad铆sticamente no significativa.

### 4. Multicolinealidad
- Se evalu贸 multicolinealidad tanto en variables num茅ricas (con Spearman) como en categ贸ricas (con Cram茅rs V), sin evidencias graves de redundancia entre predictores.

### 5. Importancia de variables (modelo predictivo)
- Al entrenar un modelo de Random Forest, se confirm贸 que las variables m谩s influyentes en la predicci贸n de enfermedad card铆aca fueron:
  - `ca`, `cp`, `oldpeak`, `thalach` y `thal`.

###  Conclusi贸n general
El an谩lisis sugiere que variables relacionadas con el sistema cardiovascular bajo esfuerzo (como `oldpeak`, `thalach`, `cp`, `exang` y `ca`) tienen un alto poder explicativo respecto al diagn贸stico de enfermedad card铆aca. Estos hallazgos pueden ser utilizados para desarrollar modelos predictivos robustos y para guiar decisiones cl铆nicas en etapas tempranas del diagn贸stico.



# Modelamiento con Random Forest

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay, roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns
import time
import pandas as pd

# Separar variables y target
X = df.drop("target", axis=1)
y = df["target"]

# Dividir en entrenamiento/validaci贸n y prueba
X_trainval, X_test, y_trainval, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Definir columnas
num_cols = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
cat_cols = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal']

# Pipelines de preprocesamiento
num_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

cat_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

preprocessor = ColumnTransformer([
    ("num", num_pipeline, num_cols),
    ("cat", cat_pipeline, cat_cols)
])

# Pipeline completo
pipeline = Pipeline([
    ("preprocessor", preprocessor),
    ("classifier", RandomForestClassifier(random_state=42))
])

# Hiperpar谩metros para GridSearch
param_grid = {
    "classifier__n_estimators": [100, 200],
    "classifier__max_depth": [None, 5, 10],
    "classifier__min_samples_split": [2, 5],
    "classifier__min_samples_leaf": [1, 2]
}

# GridSearchCV con scoring basado en recall
grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring="recall", n_jobs=-1, verbose=1)

# Entrenar y medir tiempo
start_time = time.time()
grid_search.fit(X_trainval, y_trainval)
elapsed_time = time.time() - start_time
print(f"\n憋 Tiempo de entrenamiento: {elapsed_time:.2f} segundos\n")

# Evaluar sobre conjunto de prueba
y_pred = grid_search.predict(X_test)
report = classification_report(y_test, y_pred, output_dict=True)
report_df = pd.DataFrame(report).transpose()
print("\n Reporte de clasificaci贸n (test):\n")
print(report_df)

# Matriz de confusi贸n
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap="Blues", xticklabels=['Sano', 'Enfermo'], yticklabels=['Sano', 'Enfermo'])
plt.title("Matriz de Confusi贸n")
plt.xlabel("Predicci贸n")
plt.ylabel("Real")
plt.show()

# Curva ROC
y_proba = grid_search.predict_proba(X_test)[:, 1]
fpr, tpr, _ = roc_curve(y_test, y_proba)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.2f}", color="red")
plt.plot([0, 1], [0, 1], linestyle="--", color="gray")
plt.title("Curva ROC")
plt.xlabel("Tasa de Falsos Positivos")
plt.ylabel("Tasa de Verdaderos Positivos")
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import classification_report
import pandas as pd

def obtener_reporte_df(y_true, y_pred):
    """Devuelve el classification_report como DataFrame limpio."""
    report = classification_report(y_true, y_pred, output_dict=True)
    df = pd.DataFrame(report).transpose()
    df.index.name = "Clase"
    return df.round(3)


reporte_df = obtener_reporte_df(y_test, y_pred)
print(reporte_df)
```

```python
from sklearn.metrics import recall_score
from scipy.stats import ttest_rel
import numpy as np

# Predicciones
y_train_pred = grid_search.predict(X_trainval)
y_test_pred = grid_search.predict(X_test)

# Recall global
recall_train = recall_score(y_trainval, y_train_pred)
recall_test = recall_score(y_test, y_test_pred)

print(f" Recall entrenamiento: {recall_train:.4f}")
print(f"И Recall prueba: {recall_test:.4f}")

# --- An谩lisis estad铆stico: test de diferencia entre proporciones ---
# Extraer solo los 铆ndices donde la clase verdadera es 1 (clase positiva)
train_positives = y_trainval == 1
test_positives = y_test == 1

# Aciertos individuales (1 si acert贸, 0 si no) en clase positiva
train_hits = (y_train_pred[train_positives] == 1).astype(int)
test_hits = (y_test_pred[test_positives] == 1).astype(int)

# Igualar tama帽o de muestras
np.random.seed(42)
min_len = min(len(train_hits), len(test_hits))
train_sample = np.random.choice(train_hits, size=min_len, replace=False)
test_sample = np.random.choice(test_hits, size=min_len, replace=False)

# Prueba t pareada
t_stat, p_val = ttest_rel(train_sample, test_sample)
print(f"\n Prueba t pareada en recall:")
print(f"  Estad铆stico t = {t_stat:.4f}")
print(f"  Valor p = {p_val:.4f}")

if p_val < 0.05:
    print("锔 Hay diferencia significativa en el recall entre entrenamiento y prueba.")
else:
    print(" No hay diferencia significativa en el recall entre entrenamiento y prueba.")
```
