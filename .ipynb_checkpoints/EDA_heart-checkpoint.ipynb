## 📘 Introducción

La enfermedad cardíaca es una de las principales causas de muerte a nivel mundial. Su diagnóstico temprano y preciso es fundamental para reducir la mortalidad y mejorar la calidad de vida de los pacientes. En este proyecto, realizamos un análisis exploratorio detallado del dataset de enfermedades cardíacas proveniente de UCI (University of California, Irvine), con el objetivo de identificar las variables más relevantes asociadas al diagnóstico de la condición.

Este análisis incluye:

- Exploración univariada y bivariada de variables numéricas y categóricas.
- Evaluación de distribución y normalidad de variables continuas.
- Análisis de correlación con la variable objetivo (`target`), que indica presencia (`1`) o ausencia (`0`) de enfermedad.
- Estudio de multicolinealidad entre variables predictoras.
- Evaluación estadística de asociación entre variables categóricas.
- Estimación de la importancia de características mediante un modelo de Random Forest.
- Aplicación del modelo Random Forest para realizar predicciones sobre la presencia de enfermedad.

El propósito es no solo identificar los factores que más influyen en el diagnóstico, sino también sentar las bases para futuros modelos de clasificación que permitan predecir el riesgo cardíaco con mayor precisión y respaldo estadístico.

## 🧾 Descripción de Variables

El dataset contiene variables clínicas, demográficas y de resultados de pruebas médicas que pueden influir en el diagnóstico de enfermedad cardíaca. A continuación, se describen las principales variables incluidas:

### 🔹 Variables Numéricas

- `age`: Edad del paciente (en años).
- `trestbps`: Presión arterial en reposo (en mm Hg).
- `chol`: Nivel de colesterol sérico en mg/dl.
- `thalach`: Frecuencia cardíaca máxima alcanzada.
- `oldpeak`: Depresión del segmento ST inducida por el ejercicio (comparada con el reposo).

### 🔹 Variables Categóricas

- `sex`: Sexo del paciente (0 = mujer, 1 = hombre).
- `cp`: Tipo de dolor en el pecho (0 = angina típica, 1 = angina atípica, 2 = dolor no anginoso, 3 = asintomático).
- `fbs`: Glucemia en ayunas > 120 mg/dl (1 = sí, 0 = no).
- `restecg`: Resultados del electrocardiograma en reposo (0, 1, 2).
- `exang`: Angina inducida por ejercicio (1 = sí, 0 = no).
- `slope`: Pendiente del segmento ST durante el esfuerzo (0 = descendente, 1 = plana, 2 = ascendente).
- `ca`: Número de vasos principales coloreados por fluoroscopía (0 a 3).
- `thal`: Tipo de talasemia (1 = normal, 2 = defecto fijo, 3 = defecto reversible).

### 🎯 Variable Objetivo

- `target`: Diagnóstico de enfermedad cardíaca (1 = presencia, 0 = ausencia).

```python
from scipy.stats import shapiro
import pandas as pd

# Variables numéricas del dataset
numerical_vars = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']

# Lista para guardar resultados
shapiro_results = []

# Aplicar Shapiro-Wilk a cada variable
for col in numerical_vars:
    stat, p = shapiro(df[col])
    shapiro_results.append({
        "Variable": col,
        "Estadístico Shapiro-Wilk": stat,
        "p-valor": p,
        "¿Distribución Normal?": "Sí" if p > 0.05 else "No"
    })

# Mostrar resultados como DataFrame
shapiro_df = pd.DataFrame(shapiro_results)
display(shapiro_df)
```

```python
# Añadir la columna "target" al análisis de correlación Spearman
vars_with_target = numerical_vars + ["target"]
spearman_corr_with_target = df[vars_with_target].corr(method='spearman').round(2)

# Mostrar la nueva matriz incluyendo el target
spearman_corr_with_target
```

```python
import plotly.express as px

fig_heatmap = px.imshow(spearman_corr_with_target,
                        text_auto=True,
                        color_continuous_scale="RdBu_r",  # Escala rojo-azul
                        zmin=-1, zmax=1,
                        title="Matriz de Correlación Spearman (incluyendo target)",
                        labels=dict(color="Correlación"))
fig_heatmap.update_layout(template="plotly_dark",
                          width=600, height=600)
fig_heatmap.show()
```

```python
import pandas as pd
from scipy.stats import chi2_contingency
import numpy as np

# Lista de variables categóricas en el dataset
cat_vars = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal']

# Función para calcular Cramér's V
def cramers_v(confusion_matrix):
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    return np.sqrt(phi2 / min(k - 1, r - 1))

# Evaluar cada variable categórica contra target
association_results = []
for col in cat_vars:
    confusion_matrix = pd.crosstab(df[col], df["target"])
    chi2, p, _, _ = chi2_contingency(confusion_matrix)
    cramers = cramers_v(confusion_matrix)
    association_results.append({
        "Variable": col,
        "Chi²": round(chi2, 2),
        "p-valor": round(p, 5),
        "Cramér's V": round(cramers, 3),
        "¿Asociación significativa (p < 0.05)?": "Sí" if p < 0.05 else "No"
    })

association_df = pd.DataFrame(association_results)

print("\nAsociación entre variables categóricas y `target` (Chi² + Cramér's V):\n")
print(association_df.to_string(index=False))
```

```python
# Añadir target a la matriz de Cramér's V
cramers_matrix_with_target = cramers_matrix.copy()

# Calcular Cramér's V entre target y todas las categóricas
target_cramers = []
for var in cat_vars:
    confusion_matrix = pd.crosstab(df[var], df["target"])
    cramers = cramers_v(confusion_matrix)
    target_cramers.append(round(cramers, 3))

# Agregar columna y fila de target
cramers_matrix_with_target["target"] = target_cramers
cramers_matrix_with_target.loc["target"] = target_cramers + [1.0]  # autocorrelación target-target

cramers_matrix_with_target
```

```python
import plotly.express as px

fig_cramers = px.imshow(cramers_matrix_with_target.astype(float),
                        color_continuous_scale="RdBu_r",
                        zmin=0, zmax=1,
                        title="Matriz de Asociación Categórica (Cramér's V) incluyendo target",
                        labels=dict(color="Cramér's V"))

# Si text_auto no funciona, simplemente actualiza el layout
fig_cramers.update_traces(texttemplate="%{z:.2f}", textfont_size=12)

fig_cramers.update_layout(template="plotly_dark",
                          width=700, height=700)
fig_cramers.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# Crear una copia para evitar modificar el original
df_model = df.copy()

# Codificar variables categóricas con LabelEncoder (solo porque RF no necesita OneHot)
cat_features = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal']
le_dict = {}

for col in cat_features:
    le = LabelEncoder()
    df_model[col] = le.fit_transform(df_model[col])
    le_dict[col] = le  # guardar el codificador por si lo necesitas luego

# Separar variables y target
X = df_model.drop(columns=["target"])
y = df_model["target"]

# Dividir en train y test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Entrenar Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

# Obtener importancia de características
feature_importances = pd.DataFrame({
    "Feature": X.columns,
    "Importancia": rf.feature_importances_
}).sort_values(by="Importancia", ascending=False)

# Mostrar en consola
print("\nImportancia de variables según Random Forest:\n")
print(feature_importances.to_string(index=False))
```

## ✅ Conclusiones del Análisis Descriptivo

Tras realizar un análisis exploratorio y estadístico del dataset de enfermedades cardíacas, se obtuvieron los siguientes hallazgos relevantes:

### 1. Distribución de la muestra
- La clase objetivo (`target`) está relativamente balanceada, lo que permite aplicar modelos de clasificación sin requerir técnicas agresivas de balanceo.
- Se observó una mayor prevalencia de la enfermedad en pacientes con características clínicas específicas como depresión del ST elevada (`oldpeak`) y frecuencia cardíaca máxima baja (`thalach`).

### 2. Variables numéricas
- Ninguna de las variables numéricas sigue una distribución normal según la prueba de Shapiro-Wilk, por lo que se utilizaron correlaciones de Spearman.
- Las variables más correlacionadas con la presencia de enfermedad fueron:
  - `thalach` (ρ = 0.43)
  - `oldpeak` (ρ = -0.42)
  - `age` (ρ = -0.24)

### 3. Variables categóricas
- La prueba de Chi-cuadrado y el coeficiente de Cramér’s V indicaron una fuerte asociación entre la enfermedad y las siguientes variables categóricas:
  - `thal` (V = 0.531)
  - `cp` (V = 0.519)
  - `ca` (V = 0.495)
  - `exang` (V = 0.430)
- Otras variables como `fbs` mostraron una relación estadísticamente no significativa.

### 4. Multicolinealidad
- Se evaluó multicolinealidad tanto en variables numéricas (con Spearman) como en categóricas (con Cramér’s V), sin evidencias graves de redundancia entre predictores.

### 5. Importancia de variables (modelo predictivo)
- Al entrenar un modelo de Random Forest, se confirmó que las variables más influyentes en la predicción de enfermedad cardíaca fueron:
  - `ca`, `cp`, `oldpeak`, `thalach` y `thal`.

### 📌 Conclusión general
El análisis sugiere que variables relacionadas con el sistema cardiovascular bajo esfuerzo (como `oldpeak`, `thalach`, `cp`, `exang` y `ca`) tienen un alto poder explicativo respecto al diagnóstico de enfermedad cardíaca. Estos hallazgos pueden ser utilizados para desarrollar modelos predictivos robustos y para guiar decisiones clínicas en etapas tempranas del diagnóstico.



# Modelamiento con Random Forest

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay, roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns
import time
import pandas as pd

# Separar variables y target
X = df.drop("target", axis=1)
y = df["target"]

# Dividir en entrenamiento/validación y prueba
X_trainval, X_test, y_trainval, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Definir columnas
num_cols = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
cat_cols = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal']

# Pipelines de preprocesamiento
num_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

cat_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

preprocessor = ColumnTransformer([
    ("num", num_pipeline, num_cols),
    ("cat", cat_pipeline, cat_cols)
])

# Pipeline completo
pipeline = Pipeline([
    ("preprocessor", preprocessor),
    ("classifier", RandomForestClassifier(random_state=42))
])

# Hiperparámetros para GridSearch
param_grid = {
    "classifier__n_estimators": [100, 200],
    "classifier__max_depth": [None, 5, 10],
    "classifier__min_samples_split": [2, 5],
    "classifier__min_samples_leaf": [1, 2]
}

# GridSearchCV con scoring basado en recall
grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring="recall", n_jobs=-1, verbose=1)

# Entrenar y medir tiempo
start_time = time.time()
grid_search.fit(X_trainval, y_trainval)
elapsed_time = time.time() - start_time
print(f"\n⏱️ Tiempo de entrenamiento: {elapsed_time:.2f} segundos\n")

# Evaluar sobre conjunto de prueba
y_pred = grid_search.predict(X_test)
report = classification_report(y_test, y_pred, output_dict=True)
report_df = pd.DataFrame(report).transpose()
print("\n📋 Reporte de clasificación (test):\n")
print(report_df)

# Matriz de confusión
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap="Blues", xticklabels=['Sano', 'Enfermo'], yticklabels=['Sano', 'Enfermo'])
plt.title("Matriz de Confusión")
plt.xlabel("Predicción")
plt.ylabel("Real")
plt.show()

# Curva ROC
y_proba = grid_search.predict_proba(X_test)[:, 1]
fpr, tpr, _ = roc_curve(y_test, y_proba)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.2f}", color="red")
plt.plot([0, 1], [0, 1], linestyle="--", color="gray")
plt.title("Curva ROC")
plt.xlabel("Tasa de Falsos Positivos")
plt.ylabel("Tasa de Verdaderos Positivos")
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import classification_report
import pandas as pd

def obtener_reporte_df(y_true, y_pred):
    """Devuelve el classification_report como DataFrame limpio."""
    report = classification_report(y_true, y_pred, output_dict=True)
    df = pd.DataFrame(report).transpose()
    df.index.name = "Clase"
    return df.round(3)


reporte_df = obtener_reporte_df(y_test, y_pred)
print(reporte_df)
```

```python
import joblib

# Guardar el mejor modelo encontrado con GridSearchCV
joblib.dump(grid_search.best_estimator_, "modelo_random_forest_pipeline.pkl")
```

```python
from sklearn.metrics import recall_score
from scipy.stats import ttest_rel
import numpy as np

# Predicciones
y_train_pred = grid_search.predict(X_trainval)
y_test_pred = grid_search.predict(X_test)

# Recall global
recall_train = recall_score(y_trainval, y_train_pred)
recall_test = recall_score(y_test, y_test_pred)

print(f"🔁 Recall entrenamiento: {recall_train:.4f}")
print(f"🧪 Recall prueba: {recall_test:.4f}")

# --- Análisis estadístico: test de diferencia entre proporciones ---
# Extraer solo los índices donde la clase verdadera es 1 (clase positiva)
train_positives = y_trainval == 1
test_positives = y_test == 1

# Aciertos individuales (1 si acertó, 0 si no) en clase positiva
train_hits = (y_train_pred[train_positives] == 1).astype(int)
test_hits = (y_test_pred[test_positives] == 1).astype(int)

# Igualar tamaño de muestras
np.random.seed(42)
min_len = min(len(train_hits), len(test_hits))
train_sample = np.random.choice(train_hits, size=min_len, replace=False)
test_sample = np.random.choice(test_hits, size=min_len, replace=False)

# Prueba t pareada
t_stat, p_val = ttest_rel(train_sample, test_sample)
print(f"\n📊 Prueba t pareada en recall:")
print(f"  Estadístico t = {t_stat:.4f}")
print(f"  Valor p = {p_val:.4f}")

if p_val < 0.05:
    print("⚠️ Hay diferencia significativa en el recall entre entrenamiento y prueba.")
else:
    print("✅ No hay diferencia significativa en el recall entre entrenamiento y prueba.")
```
